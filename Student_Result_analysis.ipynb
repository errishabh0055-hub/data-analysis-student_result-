```python
#DROP UNNAMED COLUMN
df=df.drop('Unnamed: 0',axis=1)
```

```python
#change wklystdyhours column
df['WklyStudyHours']=df['WklyStudyHours'].str.replace('5 - 10','5-10')
df.head()
```

```python
#GENDER COUNT
ax=sns.countplot(data=df,x='Gender',palette='pastel')
ax.bar_label(ax.containers[1])
ax.bar_label(ax.containers[0]) #easy way to count
plt.Figure(figsize=(6,6)) #easy way to changer grafe size
plt.title('Gender Distribution')
```

```python
st_p=df.groupby('ParentEduc').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(st_p)
```

```python
sns.heatmap(st_p,annot=True)                    #graf ke liye & sari value lene ke liye ,annot=true
plt.title('Parents Education & Student s Score')
```

```python
st_m=df.groupby('ParentMaritalStatus').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
print(st_m)
```

```python
sns.heatmap(st_m,annot=True)
plt.title('ParentMaritalStatus & Student s Score')
```

```python
#no impact  on the student score due to thier parent's marital status
```

```python
st_w=df.groupby('WklyStudyHours').agg({'MathScore':'mean','ReadingScore':'mean','WritingScore':'mean'})
sns.heatmap(st_w,annot=True)
plt.title('WklyStudyHours & Student s Score')
```
